```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, skew, kurtosis
```

```python
file_path = '../Lab-2/Data/salaries.csv'
data = pd.read_csv(file_path)
```

```python
salaries = data['salary_in_usd']
```

```python
# Расчет статистических параметров
mean_salary = np.mean(salaries)
variance_salary = np.var(salaries)
std_salary = np.sqrt(variance_salary)
skewness = skew(salaries)
excess_kurtosis = kurtosis(salaries)
```

```python
# Квантили
quantile_0_05 = np.quantile(salaries, 0.05)
quantile_0_95 = np.quantile(salaries, 0.95)
quantile_2_5_percent = np.quantile(salaries, 0.025)
```

```python
# Построение гистограммы исходного распределения зарплат
plt.figure(figsize=(12, 6))
count, bins, ignored = plt.hist(salaries, bins=30, density=True, alpha=0.6, color='b', edgecolor='black', label='Empirical Distribution')

# Построение нормального распределения с рассчитанными параметрами
x = np.linspace(min(salaries), max(salaries), 100)
p = norm.pdf(x, mean_salary, std_salary)
plt.plot(x, p, 'r', linewidth=2, label='Normal Distribution (fitted)')
plt.title('Comparison of Empirical Salary Distribution and Fitted Normal Distribution')
plt.xlabel('Salary in USD')
plt.ylabel('Density')
plt.legend() 
```

```python
# Печать результатов
print(f"Математическое ожидание: {mean_salary:.2f}")
print(f"Дисперсия: {variance_salary:.2f}")
print(f"Асимметрия: {skewness:.2f}")
print(f"Эксцесс: {excess_kurtosis:.2f}")
print(f"Квантиль уровня 0.05: {quantile_0_05:.2f}")
print(f"Квантиль уровня 0.95: {quantile_0_95:.2f}")
print(f"2.5%-ная точка: {quantile_2_5_percent:.2f}")
```
